```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Print all prime numbers between 0 and 50
print("Prime numbers between 0 and 50:")
for number in range(51):
    if is_prime(number):
        print(number)
```

```python
Q9.Write a program to check if a given number 371 is an Armstrong number?
```

```python
# Function to check if a number is an Armstrong number
def is_armstrong(number):
    # Convert the number to a string to iterate through each digit
    digits = str(number)
    num_digits = len(digits)
    
    # Calculate the sum of each digit raised to the power of the number of digits
    sum_of_powers = sum(int(digit) ** num_digits for digit in digits)
    
    # Check if the sum of powers is equal to the original number
    return sum_of_powers == number

# Given number
number = 371

# Check if the number is an Armstrong number
result = is_armstrong(number)

# Print the result
if result:
    print(number, "is an Armstrong number.")
else:
    print(number, "is not an Armstrong number.")
```

```python
# Function to check if a year is a leap year
def is_leap_year(year):
    # A year is a leap year if it is divisible by 4, 
    # but not divisible by 100, unless it is also divisible by 400.
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Given year
year = 1996

# Check if the year is a leap year
if is_leap_year(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
# Function to find the differences between consecutive numbers
def consecutive_differences(numbers):
    differences = []
    for i in range(len(numbers) - 1):
        diff = numbers[i + 1] - numbers[i]
        differences.append(diff)
    return differences

# Example list of numbers
numbers_list = [10, 15, 25, 30, 50]

# Calculate the differences
result = consecutive_differences(numbers_list)

# Print the resulting differences
print("Differences between consecutive numbers:", result)
```

```python
Q23. Write a function called word_count(text) that takes a string as input and returns a
dictionary where Each key is a word in the text and its value is the number of times that word appears in the text.
For example, word_count("hello world hello") should return {"hello": 2, "world": 1}.
```

```python
def word_count(text):
    # Split the text into words
    words = text.split()
    
    # Create a dictionary to hold word counts
    count_dict = {}
    
    # Count each word's occurrences
    for word in words:
        if word in count_dict:
            count_dict[word] += 1
        else:
            count_dict[word] = 1
            
    return count_dict

# Test the function
result = word_count("hello world hello")
print(result)  # Output: {'hello': 2, 'world': 1}
```

```python
Q33. Write a Python class called "BankAccount" with attributes for "balance" and "interest_rate".
Implement methods to deposit and withdraw money from the account, as well as to calculate the interest
earned on the balance based on the interest rate. Create an instance of this class and use it to test out the
implemented methods for deposit, withdrawal, and interest calculation.
```

```python
# Define the BankAccount class
class BankAccount:
    def __init__(self, initial_balance, interest_rate):
        self.balance = initial_balance
        self.interest_rate = interest_rate

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Withdrawal amount must be positive and cannot exceed the balance.")

    def calculate_interest(self):
        interest = self.balance * (self.interest_rate / 100)
        return interest

# Create an instance of the BankAccount class
account = BankAccount(1000, 5)  # Initial balance of $1000 and 5% interest rate

# Test the deposit method
account.deposit(200)

# Test the withdraw method
account.withdraw(150)

# Calculate and print the interest earned
interest_earned = account.calculate_interest()
print(f"Interest earned: ${interest_earned:.2f}")
```

```python
import math

# Define the Shape class
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement this method")

    def perimeter(self):
        raise NotImplementedError("Subclasses must implement this method")

# Define the Rectangle subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# Define the Circle subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.radius

# Define the Triangle subclass
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self, side1, side2):
        return self.base + side1 + side2

# Create instances of each shape
rectangle = Rectangle(4, 5)
circle = Circle(3)
triangle = Triangle(6, 4)

# For the triangle, we need to provide side lengths for perimeter calculation
triangle_side1 = 5
triangle_side2 = 7

# Print out the area and perimeter of each shape
print("Rectangle - Area:", rectangle.area(), "Perimeter:", rectangle.perimeter())
print("Circle - Area:", circle.area(), "Perimeter:", circle.perimeter())
print("Triangle - Area:", triangle.area(), "Perimeter:", triangle.perimeter(triangle_side1, triangle_side2))
```

```python
import csv

def calculate_average_of_column(file_path, column_index):
    """Calculates the average of a specific column in a CSV file."""
    total = 0
    count = 0
    
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)  # Skip the header row if there is one
            
            for row in reader:
                if row:  # Check if the row is not empty
                    try:
                        value = float(row[column_index])  # Convert the column value to float
                        total += value
                        count += 1
                    except (ValueError, IndexError):
                        # Handle cases where the conversion fails or index is out of range
                        print(f"Skipping invalid value in row: {row}")
            
            if count == 0:
                print("No valid values found in the specified column.")
                return None
            
            average = total / count
            return average
            
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
file_path = 'example.csv'  # Replace with your CSV file path
column_index = 1  # Specify the index of the column you want to average (0-based index)
average = calculate_average_of_column(file_path, column_index)

if average is not None:
    print(f"The average of column {column_index} in '{file_path}' is: {average:.2f}")
```

```python
Q62. Write a program that generates a list of random numbers and sorts them using multithreading. Use
two threads to sort the first half and the second half of the list in parallel.
```

```python
import threading
import random

# Function to sort a portion of the list
def sort_portion(portion, result, index):
    result[index] = sorted(portion)  # Sort the portion and store it in the result list

# Generate a list of random numbers
def generate_random_numbers(size):
    return [random.randint(1, 100) for _ in range(size)]

# Main function
def main():
    size = 10  # Size of the list
    numbers = generate_random_numbers(size)  # Generate random numbers
    print("Original list:", numbers)

    # Split the list into two halves
    mid = size // 2
    first_half = numbers[:mid]
    second_half = numbers[mid:]

    # Prepare a list to hold sorted results
    sorted_result = [None, None]

    # Create threads for sorting
    thread1 = threading.Thread(target=sort_portion, args=(first_half, sorted_result, 0))
    thread2 = threading.Thread(target=sort_portion, args=(second_half, sorted_result, 1))

    # Start the threads
    thread1.start()
    thread2.start()

    # Wait for both threads to finish
    thread1.join()
    thread2.join()

    # Merge the sorted halves
    sorted_numbers = sorted_result[0] + sorted_result[1]
    print("Sorted list:", sorted(sorted_numbers))

if __name__ == "__main__":
    main()
```

```python
import threading
import time

class BankAccount:
    def __init__(self):
        self.balance = 0
        self.lock = threading.Lock()  # To prevent race conditions

    def deposit(self, amount):
        with self.lock:
            print(f"Depositing {amount}. Current balance: {self.balance}")
            self.balance += amount
            print(f"New balance after deposit: {self.balance}")

    def withdraw(self, amount):
        with self.lock:
            if self.balance >= amount:
                print(f"Withdrawing {amount}. Current balance: {self.balance}")
                self.balance -= amount
                print(f"New balance after withdrawal: {self.balance}")
            else:
                print(f"Withdrawal of {amount} failed. Current balance: {self.balance}")

def deposit_thread(account):
    for _ in range(10):
        account.deposit(100)
        time.sleep(0.1)  # Adding a slight delay to simulate processing time

def withdraw_thread(account):
    for _ in range(10):
        account.withdraw(50)
        time.sleep(0.1)  # Adding a slight delay to simulate processing time

if __name__ == "__main__":
    account = BankAccount()

    # Create threads
    deposit_thread_instance = threading.Thread(target=deposit_thread, args=(account,))
    withdraw_thread_instance = threading.Thread(target=withdraw_thread, args=(account,))

    # Start threads
    deposit_thread_instance.start()
    withdraw_thread_instance.start()

    # Wait for both threads to finish
    deposit_thread_instance.join()
    withdraw_thread_instance.join()

    print(f"Final balance: {account.balance}")
```

```python
import multiprocessing

def compute_partial_sum(start, end, result, index):
    """Compute the sum of integers from start to end (inclusive)."""
    partial_sum = sum(range(start, end + 1))
    result[index] = partial_sum

if __name__ == "__main__":
    n = 1000
    num_processes = 4
    chunk_size = n // num_processes

    # Create a shared array to store results
    result = multiprocessing.Array('i', num_processes)

    # Create processes
    processes = []
    for i in range(num_processes):
        start = i * chunk_size + 1
        end = (i + 1) * chunk_size if i < num_processes - 1 else n
        p = multiprocessing.Process(target=compute_partial_sum, args=(start, end, result, i))
        processes.append(p)
        p.start()

    # Wait for all processes to finish
    for p in processes:
        p.join()

    # Calculate total sum from results
    total_sum = sum(result)
    print(f"The sum of the first {n} integers is: {total_sum}")
```

```python
Q80.Write a Python program that uses the multiprocessing module to parallelize the computation of the factorial of 
a number.
```

```python
import multiprocessing
import math

def factorial_segment(start, end):
    """Compute the product of numbers from start to end (inclusive)."""
    result = 1
    for i in range(start, end + 1):
        result *= i
    return result

def parallel_factorial(n):
    """Calculate the factorial of n using multiple processes."""
    if n == 0 or n == 1:
        return 1
    
    # Determine the number of processes
    num_processes = multiprocessing.cpu_count()
    
    # Split the work into segments
    segment_size = n // num_processes
    segments = [(i * segment_size + 1, (i + 1) * segment_size) for i in range(num_processes)]
    
    # Handle the last segment, which may be larger
    segments[-1] = (segments[-1][0], n)
    
    # Create a Pool of worker processes
    with multiprocessing.Pool(processes=num_processes) as pool:
        # Map the segments to the worker processes
        results = pool.starmap(factorial_segment, segments)

    # Calculate the final factorial by multiplying the results
    final_result = math.prod(results)
    return final_result

if __name__ == "__main__":
    number = 10  # Change this to any positive integer to compute its factorial
    result = parallel_factorial(number)
    print(f"The factorial of {number} is: {result}")
```

```python
from pymongo import MongoClient
from pymongo.errors import PyMongoError

def connect_to_mongo(db_name):
    """Connect to MongoDB and return the database."""
    try:
        client = MongoClient('localhost', 27017)  # Change host and port if needed
        print("Successfully connected to MongoDB.")
        return client[db_name]
    except PyMongoError as e:
        print(f"Error connecting to MongoDB: {e}")
        return None

def aggregate_data(collection):
    """Perform aggregation to summarize user data."""
    try:
        # Define the aggregation pipeline
        pipeline = [
            {
                "$group": {
                    "_id": "$age",  # Group by age
                    "count": {"$sum": 1}  # Count the number of users in each age group
                }
            },
            {
                "$sort": {"_id": 1}  # Sort by age in ascending order
            }
        ]

        # Execute the aggregation pipeline
        results = collection.aggregate(pipeline)

        # Display the results
        print("Age | Count")
        print("-------------")
        for doc in results:
            print(f"{doc['_id']}  | {doc['count']}")

    except PyMongoError as e:
        print(f"Error during aggregation: {e}")

def main():
    # Connect to the MongoDB database
    db_name = 'my_database'  # Replace with your database name
    db = connect_to_mongo(db_name)

    if db:
        # Select the collection
        collection = db['users']  # Replace with your collection name

        # Perform aggregation
        aggregate_data(collection)

        # Close the connection
        print("MongoDB connection closed.")

if __name__ == "__main__":
    main()
```
