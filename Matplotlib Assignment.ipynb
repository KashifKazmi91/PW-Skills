````python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
````

````python
Matplotlib is a popular plotting library for the Python programming language and its numerical mathematics extension, 
NumPy. It provides a flexible and powerful framework for creating static, animated, and interactive visualizations in 
Python.

Why is it Used?
- Data Visualization: Matplotlib is used to visualize data in a variety of formats, making it easier to understand 
patterns, trends, and relationships in the data.
- Customizable Plots: It offers a high degree of customization, allowing users to modify the appearance of plots 
(e.g., colors, labels, scales).
- Wide Range of Plots: It can generate various types of plots, making it suitable for different data types and analysis
needs.
- Integration: Matplotlib integrates well with other libraries like NumPy, pandas, and Jupyter notebooks, making it a 
versatile tool for data analysis.

Five Plots that can be Plotted Using the Pyplot Module:
1. Line Plot: Used to visualize data points connected by lines.
   ```python
   plt.plot(x, y)
   ```

2. Bar Plot: Used for comparing different categories with rectangular bars.
   ```python
   plt.bar(x, height)
   ```

3. Histogram: Used to show the distribution of numerical data.
   ```python
   plt.hist(data, bins)
   ```

4. Scatter Plot: Used to display values for typically two variables for a set of data.
   ```python
   plt.scatter(x, y)
   ```

5. Pie Chart: Used to represent proportions of a whole in a circular graph.
   ```python
   plt.pie(sizes, labels)
   ```

These plots help in visualizing data effectively for analysis and presentations.
````

````python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title("Scatter Plot Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
````

````python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data for the line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Four different y values for each line
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure
plt.figure(figsize=(10, 8))

# Create 4 subplots (2 rows and 2 columns)
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1', color='b')
plt.title('Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2', color='g')
plt.title('Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3', color='r')
plt.title('Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4', color='m')
plt.title('Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

````python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
````

````python
Why is it Used?
Comparison: Bar plots effectively compare different groups or categories side by side.
Categorical Data: They are particularly useful for visualizing categorical data where you want to show the frequency 
or amount associated with each category.
Simplicity: Bar plots are straightforward and easy to interpret, making them a popular choice for presenting data.
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(10, 5))

# Vertical Bar Plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.bar(company, profit, color='blue')
plt.title('Company Profit (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid(axis='y')

# Horizontal Bar Plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.barh(company, profit, color='orange')
plt.title('Company Profit (Horizontal Bar Plot)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.grid(axis='x')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

````python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot of Box 1 and Box 2')
plt.xlabel('Boxes')
plt.ylabel('Values')

# Show the plot
plt.grid(axis='y')
plt.show()
````
