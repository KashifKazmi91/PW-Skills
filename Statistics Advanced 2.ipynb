```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.
```

```python
Probability Mass Function (PMF)

The Probability Mass Function (PMF) is used for discrete random variables. It gives the probability that a discrete 
random variable is equal to a specific value.

Example of PMF:
Consider a fair six-sided die. The PMF for the outcome of rolling the die can be expressed as:
- (P(X = 1) = {1}/{6})
- (P(X = 2) = {1}/{6})
- (P(X = 3) = {1}/{6})
- (P(X = 4) = {1}/{6})
- (P(X = 5) = {1}/{6})
- (P(X = 6) = {1}/{6})

Here, each outcome has an equal probability of ({1}/{6}).

Probability Density Function (PDF)

The Probability Density Function (PDF) is used for continuous random variables. It describes the likelihood of a 
random variable to take on a particular value. Unlike the PMF, the PDF does not give probabilities directly but rather 
densities, and the probability of the variable falling within a certain range is given by the area under the curve of
the PDF over that range.

Example of PDF:
Consider a continuous random variable representing the height of adult men, which may follow a normal distribution. 
The PDF for this distribution might look something like this:
- The mean height might be 175 cm, with a standard deviation of 10 cm. 
- The PDF would be shaped like a bell curve centered at 175 cm, with the area under the curve representing the 
probabilities of different height ranges.
```

```python
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
```

```python
Cumulative Distribution Function (CDF)
The Cumulative Distribution Function (CDF) of a random variable gives the probability that the variable takes on a 
value less than or equal to a specific point. It is a function that provides the cumulative probability for a random 
variable, whether it is discrete or continuous.

Example of CDF
1. Discrete Example:
   Consider a six-sided die. The PMF for rolling the die is:

- (P(X = 1) = {1}/{6})
- (P(X = 2) = {1}/{6})
- (P(X = 3) = {1}/{6})
- (P(X = 4) = {1}/{6})
- (P(X = 5) = {1}/{6})
- (P(X = 6) = {1}/{6})

The CDF for this die would be calculated as follows:

- (F(1) = P(X leq 1) = {1}/{6})
- (F(2) = P(X leq 2) = {1}/{6} + {1}/{6} = {2}/{6} = {1}/{3})
- (F(3) = P(X leq 3) = {3}/{6} = {1}/{2})
- (F(4) = P(X leq 4) = {4}/{6} = {2}/{3})
- (F(5) = P(X leq 5) = {5}/{6})
- (F(6) = P(X leq 6) = 1)

The CDF increases stepwise, with the maximum value of 1 at (X = 6).

2. Continuous Example:
   For a continuous random variable, such as the height of individuals following a normal distribution with mean (mu) 
and standard deviation (sigma).
```

```python
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
```

```python
The normal distribution is a fundamental probability distribution in statistics and is often used as a model in 
various real-world situations. Here are some examples where the normal distribution might be applied:

Examples of Situations
1. Heights and Weights:
   - Human heights and weights often follow a normal distribution. For instance, if you measure the heights of a 
large group of adults, you might find that most people are near the average height, with fewer individuals being 
very short or very tall.

2. Test Scores:
   - Standardized test scores (like SAT or IQ tests) are typically modeled as normal distributions. Most students 
score near the average, while fewer students score very low or very high.

3. Measurement Errors:
   - In scientific experiments, measurement errors often follow a normal distribution. If you measure the same quantity
multiple times, the variations in your measurements will likely be normally distributed around the true value.

4. Financial Returns:
   - The daily returns on stocks and other financial assets are often assumed to be normally distributed. This helps 
in modeling risks and making investment decisions.

5. Quality Control:
   - In manufacturing, the dimensions of products (like the diameter of a bolt) are often normally distributed around 
a target value, with specifications set to control how much variation is acceptable.

Parameters of the Normal Distribution
The normal distribution is defined by two key parameters:
1. Mean:
   - The mean is the center of the distribution. It determines the location of the peak of the bell curve. In a normal 
distribution, the mean divides the data into two equal halves. Changing the mean shifts the entire distribution left or
right along the x-axis.

2. Standard Deviation:
   - The standard deviation measures the spread or dispersion of the distribution. A smaller standard deviation results
in a steeper and narrower curve, while a larger standard deviation produces a flatter and wider curve. It indicates how
much the values typically deviate from the mean.
```

```python
Importance of Normal Distribution
The normal distribution is one of the most important concepts in statistics due to several key reasons:
1. Central Limit Theorem:
   - The central limit theorem states that the sum (or average) of a large number of independent random variables, 
regardless of their original distribution, will tend to be normally distributed. This property makes the normal 
distribution applicable in a wide range of real-world scenarios.
2. Statistical Inference:
   - Many statistical methods, including hypothesis testing and confidence intervals, assume normality. This allows 
researchers to make inferences about populations based on sample data.
3. Simplicity and Predictability:
   - The normal distribution is mathematically tractable, making it easier to work with. It has well-defined properties,
which allows for easy calculation of probabilities and percentiles.
4. Modeling Real-World Phenomena:
   - Many natural and social phenomena tend to cluster around a mean, making the normal distribution a good fit for 
modeling these behaviors.

Real-Life Examples of Normal Distribution
1. Human Characteristics:
   - Traits such as height, weight, and IQ scores often follow a normal distribution. For example, if you measure the 
heights of adult men, the majority will cluster around the average height, with fewer individuals being very short or 
very tall.
2. Test Scores:
   - Standardized test scores (e.g., SAT, GRE) typically have a normal distribution. Most students score near the 
average, while very few achieve extremely low or high scores.
3. Measurement Errors:
   - In scientific experiments, errors in measurements often follow a normal distribution. If a scientist measures the 
same quantity multiple times, the variations in those measurements will typically be normally distributed around the
true value.
4. Financial Returns:
   - Daily returns of stocks and other financial assets are often modeled using a normal distribution. While the actual 
returns may not be perfectly normal due to market volatility, many financial models assume normality for ease of 
calculation.
5. Manufacturing Quality Control:
   - In quality control processes, product measurements (e.g., dimensions, weight) are often assumed to be normally 
distributed around a target value. This helps in setting acceptable tolerance levels and determining the percentage 
of products that meet quality standards.
6. Natural Phenomena:
   - Measurements such as the duration of a specific event (e.g., lifespan of light bulbs, the time taken to complete 
a task) can often be modeled using a normal distribution, as they tend to cluster around a mean with variability.
```

```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
```

```python
Bernoulli Distribution
The Bernoulli distribution is a discrete probability distribution that describes a random experiment with exactly 
two possible outcomes: "success" (often coded as 1) and "failure" (coded as 0). It is characterized by a single 
parameter \(p\), which represents the probability of success.

Example of Bernoulli Distribution

Consider a simple experiment of flipping a coin:
- Let "heads" represent success (1) and "tails" represent failure (0).
- If the coin is fair, then (p = 0.5) for heads (success) and (1 - p = 0.5) for tails (failure).

Thus, the Bernoulli distribution for this experiment can be described as:
- (P(X = 1) = 0.5)
- (P(X = 0) = 0.5)

Difference Between Bernoulli Distribution and Binomial Distribution

1. Definition:
   - Bernoulli Distribution: Describes a single trial with two possible outcomes.
   - Binomial Distribution: Describes the number of successes in (n) independent Bernoulli trials 
(where each trial has the same probability of success (p)).

2. Parameters:
   - Bernoulli Distribution: Has a single parameter (p) (the probability of success).
   - Binomial Distribution: Has two parameters: (n) (the number of trials) and (p) (the probability of success).

3. Outcome:
   - Bernoulli Distribution: Can result in either 0 or 1.
   - Binomial Distribution: Can result in any integer from 0 to (n) (the number of successes in (n) trials).
```

```python
Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.
```

```python
To calculate the probability that a randomly selected observation from a normally distributed dataset with a mean of 50
and a standard deviation of 10 is greater than 60, we can use the Z-score formula and then find the corresponding 
probability using the standard normal distribution.

Step 1: Calculate the Z-score
The Z-score is calculated using the formula:

Z = {X - mu}/{sigma}

Where:
- (X) is the value we are interested in (in this case, 60),
- (mu) is the mean (50),
- (sigma) is the standard deviation (10).

Substituting the values into the formula:

Z = {60 - 50}/{10} = {10}/{10} = 1

Step 2: Find the Probability

Now, we need to find the probability that (Z > 1). This requires looking up the Z-score in the standard normal 
distribution table or using a cumulative distribution function (CDF) calculator.

- The CDF gives us the probability that a random variable is less than or equal to a certain value. For (Z = 1), 
the CDF is approximately 0.8413. This means that approximately 84.13% of the observations fall below 60.

Step 3: Calculate the Probability Greater Than 60

To find the probability that a randomly selected observation is greater than 60, we can use:

P(X > 60) = 1 - P(X \leq 60) = 1 - P(Z \leq 1)

Substituting the value we found:

P(X > 60) = 1 - 0.8413 = 0.1587
```

```python
Uniform Distribution
The uniform distribution is a type of probability distribution where all outcomes are equally likely to occur. 
In other words, each value within a specified range has the same probability of being selected. There are two types 
of uniform distributions:

1. Discrete Uniform Distribution: This applies to scenarios where there are a finite number of possible outcomes.
2. Continuous Uniform Distribution: This applies to scenarios where outcomes can take any value within a continuous 
interval.

Example: Rolling a fair six-sided die.
In this case:
- The possible outcomes are {1, 2, 3, 4, 5, 6}.
- Each outcome has a probability of ( {1}/{6} ) since all outcomes are equally likely.

The probability mass function (PMF) for this discrete uniform distribution can be expressed as:
P(X = x) = {1}/{n}

Where (n) is the number of possible outcomes (in this case, 6).

Example: Selecting a random number between 0 and 1.

In this case:
- The possible outcomes are any real number (x) in the interval ([0, 1]).
- The probability density function (PDF) for a continuous uniform distribution is defined as:

f(x) = begin{cases} \frac{1}{b - a} & \text{for } a \leq x \leq b \\ 0 & \text{otherwise}  end{cases}
```

```python
Q8: What is the z score? State the importance of the z score.
```

```python
Z Score
The Z score, also known as the standard score, measures how many standard deviations a data point (or observation) is 
from the mean of a dataset. It is calculated using the following formula:

Z = {X - mu}/{sigma}

Where:
- (Z) is the Z score,
- (X) is the value of the observation,
- (mu) is the mean of the dataset,
- (sigma) is the standard deviation of the dataset.

Importance of the Z Score

1. Standardization:
   - Z scores standardize different datasets, allowing for comparison between data points from different distributions.
This is particularly useful when dealing with different units or scales.

2. Identifying Outliers:
   - Z scores help identify outliers in a dataset. A Z score greater than 3 or less than -3 typically indicates that a 
data point is significantly different from the rest of the data.

3. Probability Calculation:
   - Z scores are used to calculate probabilities and percentiles in a standard normal distribution. They allow 
researchers to determine how likely it is for a data point to fall above or below a certain value.

4. Statistical Inference:
   - In hypothesis testing, Z scores are crucial for determining whether to reject or fail to reject the null 
hypothesis based on the significance level.

5. Normal Distribution:
   - Z scores are integral to the properties of the normal distribution. By converting raw scores to Z scores, one can 
utilize standard normal distribution tables to find probabilities and critical values.

6. Comparing Different Samples:
   - Z scores enable comparison across different samples or populations, helping to assess performance or outcomes
relative to a common scale.
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
```

```python
Central Limit Theorem (CLT)
The Central Limit Theorem states that, given a sufficiently large sample size, the sampling distribution of the sample
mean (or sum) will be approximately normally distributed, regardless of the shape of the population distribution from 
which the samples are drawn. This approximation improves as the sample size increases.

Key Points of the CLT:
1. Sample Mean: The theorem primarily concerns the distribution of the sample mean.
2. Sufficiently Large Sample Size: Typically, a sample size of 30 or more is considered adequate for the CLT to hold 
true.
3. Independence: The samples must be independent of each other.
4. Population Distribution: The population can have any distribution (normal, skewed, etc.).

Significance of the Central Limit Theorem
1. Foundation for Inferential Statistics:
   - The CLT provides the theoretical foundation for many statistical methods, enabling us to make inferences about 
population parameters based on sample statistics.
2. Normal Approximation:
   - It allows researchers to use the normal distribution as an approximation for the sampling distribution of the 
mean, simplifying calculations and statistical tests.
3. Hypothesis Testing:
   - The CLT justifies the use of z-tests and t-tests, allowing statisticians to test hypotheses about population 
means and proportions.
4. Confidence Intervals:
   - It underpins the construction of confidence intervals, providing a method to estimate the uncertainty around 
sample estimates.
5. Handling Non-Normal Populations:
   - The CLT applies even when the population distribution is not normal, making it a powerful tool in practical data 
analysis scenarios.
6. Predictability:
   - It assures that, as sample sizes increase, the variability of the sample mean decreases, leading to more 
predictable and stable estimates.
```

```python
Q10: State the assumptions of the Central Limit Theorem.
```

```python
The Central Limit Theorem (CLT) relies on several key assumptions to hold true. These assumptions ensure that the 
sampling distribution of the sample mean approaches a normal distribution as the sample size increases. Here are the 
primary assumptions of the CLT:

Assumptions of the Central Limit Theorem
1. Independence:
   - The sampled observations must be independent of one another. This means that the selection of one observation 
should not affect the selection of another.
2. Random Sampling:
   - The samples must be drawn randomly from the population. This ensures that each member of the population has an 
equal chance of being included in the sample.
3. Sample Size:
   - A sufficiently large sample size is required for the CLT to hold. While there isn't a strict cutoff, a common rule 
of thumb is that a sample size of 30 or more is typically considered adequate. However, if the population distribution 
is heavily skewed or has outliers, a larger sample size may be necessary.
4. Finite Mean and Variance:
   - The population from which the samples are drawn must have a finite mean (mu) and a finite variance (sigma^2). 
Infinite means or variances (as in some heavy-tailed distributions) can violate the assumptions of the CLT.
5. Shape of the Population Distribution:
   - The population distribution can be normal or non-normal. The CLT assures that the sampling distribution of the 
sample mean will approximate a normal distribution as the sample size increases, regardless of the shape of the 
population distribution.
```
